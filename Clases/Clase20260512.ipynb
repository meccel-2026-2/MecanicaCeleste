## 1. Definición analítica de un funcional

En el análisis matemático convencional, una **función** $f(x)$ es una regla de correspondencia que asigna a un número real $x$ (o un vector) otro número real. En cambio, un **funcional** es un objeto matemático que toma una **función completa** como argumento y devuelve un número escalar.

Formalmente, si tenemos un conjunto de funciones admisibles $\mathcal{Y}$, un funcional $J$ es un mapeo de la forma:

$$J: \mathcal{Y} \to \mathbb{R}$$

En física, los funcionales suelen expresarse como integrales definidas donde el integrando depende de una función $y(x)$, su derivada $y'(x)$ y la variable independiente $x$:

$$J[y] = \int_{x_1}^{x_2} F(x, y, y') dx$$

Aquí, el valor de $J$ depende de la "forma" de la curva $y(x)$ entre los límites $x_1$ y $x_2$. Si cambiamos la trayectoria de la función, el valor del funcional cambia.

---

## 2. Ejemplo: El problema de la braquistócrona o la distancia mínima

El ejemplo más intuitivo es el cálculo de la **longitud de arco** entre dos puntos fijos $A$ y $B$. Queremos encontrar la función $y(x)$ que minimice la distancia total.

La longitud infinitesimal de una curva en el plano es $ds = \sqrt{dx^2 + dy^2} = \sqrt{1 + (y')^2} dx$. El funcional de longitud total es:

$$L[y] = \int_{x_A}^{x_B} \sqrt{1 + \left(\frac{dy}{dx}\right)^2} dx$$

En este caso, $F(x, y, y') = \sqrt{1 + (y')^2}$. Resolver el problema consiste en hallar qué función $y(x)$ hace que este funcional sea un extremo (mínimo).

---

## 3. Deducción de las ecuaciones de Lagrange

Para deducir las ecuaciones que rigen estos extremos, partimos de un funcional general:

$$J[y] = \int_{x_1}^{x_2} F(x, y, y') dx$$

### El método de variaciones

Supongamos que $y(x)$ es la función exacta que hace que $J[y]$ sea un extremo. Si introducimos una pequeña deformación o "variación" sobre esta trayectoria óptima, el valor del funcional no debería cambiar (en primera aproximación).

Definimos una familia de funciones cercanas:


$$y(x, \epsilon) = y(x) + \epsilon \eta(x)$$

Donde:

* $\epsilon$ es un parámetro infinitesimal.
* $\eta(x)$ es una función de prueba arbitraria y diferenciable que debe anularse en los extremos, es decir, $\eta(x_1) = \eta(x_2) = 0$, para asegurar que todas las curvas comparadas pasen por los mismos puntos fijos.

### Diferenciación del funcional

Al sustituir la función variada en el funcional, $J$ se convierte en una función de $\epsilon$:


$$J(\epsilon) = \int_{x_1}^{x_2} F(x, y + \epsilon \eta, y' + \epsilon \eta') dx$$

Para que $y(x)$ sea un extremo, la derivada de $J$ respecto a $\epsilon$ debe ser cero cuando $\epsilon \to 0$:


$$\frac{dJ}{d\epsilon} \bigg|_{\epsilon=0} = 0$$

Aplicando la regla de la cadena dentro de la integral:


$$\int_{x_1}^{x_2} \left( \frac{\partial F}{\partial y} \frac{\partial (y + \epsilon \eta)}{\partial \epsilon} + \frac{\partial F}{\partial y'} \frac{\partial (y' + \epsilon \eta')}{\partial \epsilon} \right) dx = 0$$

Evaluando en $\epsilon = 0$:


$$\int_{x_1}^{x_2} \left( \frac{\partial F}{\partial y} \eta(x) + \frac{\partial F}{\partial y'} \eta'(x) \right) dx = 0$$

### Integración por partes

El segundo término contiene $\eta'(x)$. Para factorizar $\eta(x)$, aplicamos integración por partes en el término $\int \frac{\partial F}{\partial y'} \eta' dx$:

* Sea $u = \frac{\partial F}{\partial y'}$ y $dv = \eta'(x) dx$.
* Entonces $du = \frac{d}{dx} \left( \frac{\partial F}{\partial y'} \right) dx$ y $v = \eta(x)$.

La integral del segundo término queda:


$$\left[ \frac{\partial F}{\partial y'} \eta(x) \right]_{x_1}^{x_2} - \int_{x_1}^{x_2} \frac{d}{dx} \left( \frac{\partial F}{\partial y'} \right) \eta(x) dx$$

Como establecimos que $\eta(x_1) = \eta(x_2) = 0$, el término de frontera desaparece. Sustituyendo esto en la ecuación principal obtenemos:


$$\int_{x_1}^{x_2} \left( \frac{\partial F}{\partial y} - \frac{d}{dx} \frac{\partial F}{\partial y'} \right) \eta(x) dx = 0$$

### El Lema Fundamental del Cálculo de Variaciones

Dado que esta integral debe ser cero para **cualquier** función arbitraria $\eta(x)$, la única forma de garantizar esto es que el integrando entre paréntesis sea idénticamente nulo. Así obtenemos la **Ecuación de Euler-Lagrange**:

$$\frac{\partial F}{\partial y} - \frac{d}{dx} \left( \frac{\partial F}{\partial y'} \right) = 0$$

## 4. Aplicación a la Mecánica Clásica

En física, el funcional fundamental es la **Acción** ($S$), que se define como la integral temporal del **Lagrangiano** ($L$):


$$S[q] = \int_{t_1}^{t_2} L(t, q, \dot{q}) dt$$

Donde:

* $q$ representa las coordenadas generalizadas.
* $\dot{q}$ es la velocidad generalizada.
* $L = T - V$ (Energía cinética menos energía potencial).

Siguiendo la deducción anterior, el **Principio de Hamilton** (o Principio de Mínima Acción) establece que la trayectoria real de un sistema es aquella que hace que $S$ sea estacionaria. Por lo tanto, sustituyendo $x \to t$, $y \to q$ y $F \to L$, obtenemos las **Ecuaciones de Lagrange**:

$$\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{q}_i} \right) - \frac{\partial L}{\partial q_i} = 0$$

Esta formulación es extremadamente potente porque es independiente del sistema de coordenadas y permite tratar ligaduras de forma mucho más elegante que las leyes de Newton tradicionales.

## 1. Desarrollo Analítico

Consideremos el funcional de acción (o longitud) definido por:


$$S[f] = \int_{x_1}^{x_2} \sqrt{1 + \left( \frac{df}{dx} \right)^2} dx$$


Donde definimos nuestro Lagrangiano como $L(f, \dot{f}, x) = \sqrt{1 + \dot{f}^2}$.

### Paso A: Identificación de componentes

Para aplicar las ecuaciones de Euler-Lagrange, calculamos las derivadas necesarias:

1. **Dependencia posicional:** Como $L$ no contiene explícitamente a $f$:

$$\frac{\partial L}{\partial f} = 0$$


2. **Momento conjugado:** Derivamos respecto a la "velocidad" $\dot{f}$:

$$\frac{\partial L}{\partial \dot{f}} = \frac{\partial}{\partial \dot{f}} (1 + \dot{f}^2)^{1/2} = \frac{1}{2}(1 + \dot{f}^2)^{-1/2} (2\dot{f})$$


$$\frac{\partial L}{\partial \dot{f}} = \frac{\dot{f}}{\sqrt{1 + \dot{f}^2}}$$



### Paso B: Ecuación de movimiento

Sustituimos en $\frac{d}{dx} \left( \frac{\partial L}{\partial \dot{f}} \right) - \frac{\partial L}{\partial f} = 0$:


$$\frac{d}{dx} \left( \frac{\dot{f}}{\sqrt{1 + \dot{f}^2}} \right) = 0$$

Esto nos dice que el término entre paréntesis es una constante, llamémosla $k$:


$$\frac{\dot{f}}{\sqrt{1 + \dot{f}^2}} = k$$

### Paso C: Resolución de la trayectoria

Para que esa fracción sea constante, la propia $\dot{f}$ debe ser constante. Si elevamos al cuadrado para demostrarlo:


$$\dot{f}^2 = k^2(1 + \dot{f}^2) \implies \dot{f}^2(1 - k^2) = k^2 \implies \dot{f} = \sqrt{\frac{k^2}{1 - k^2}} = \text{Constante}$$

Al integrar una velocidad constante respecto a $x$, obtenemos la ecuación de una **línea recta**:


$$f(x) = mx + b$$

## 2. Explicación Conceptual en Mecánica Clásica

¿Por qué es importante este desarrollo si solo nos da una línea recta? La respuesta reside en la transición de la física de fuerzas a la física de principios variacionales.

### I. El Principio de Inercia de Newton

En mecánica clásica, la Primera Ley de Newton dice que un cuerpo libre de fuerzas se mueve en línea recta a velocidad constante. El desarrollo de este Lagrangiano es la **prueba matemática** de esa ley desde una perspectiva geométrica:

* Newton dice: "No hay fuerzas, por tanto no hay aceleración".
* Lagrange dice: "La naturaleza minimiza la distancia, y en un espacio plano, la distancia mínima es la línea recta".

### II. Geometrización de la Física

Este Lagrangiano trata a la coordenada $f$ y a la variable $x$ de forma puramente geométrica. Nos dice que el sistema físico se comporta como una **geodésica**. En mecánica clásica avanzada, esto es fundamental para entender que:

> Las partículas no "sienten fuerzas", sino que simplemente siguen el camino más corto (o extremo) en la configuración del espacio en el que se encuentran.

### III. El límite de "Acción Mínima"

Si expandimos este Lagrangiano mediante una serie de Taylor para valores de $\dot{f}$ pequeños (como hacemos al aproximar funciones en física):


$$\sqrt{1 + \dot{f}^2} \approx 1 + \frac{1}{2}\dot{f}^2 - \frac{1}{8}\dot{f}^4 + \dots$$


Observamos que, salvo la constante inicial, el término dominante es **$\frac{1}{2}\dot{f}^2$**.
Este es precisamente el término de **energía cinética** ($T = \frac{1}{2}mv^2$) que usamos en el Lagrangiano estándar $L = T - V$.

Por lo tanto, $L = \sqrt{1 + \dot{f}^2}$ es la versión "geométricamente pura" de la energía cinética. Mientras que el Lagrangiano estándar $T-V$ es una aproximación energética, este Lagrangiano es una declaración sobre la estructura métrica del espacio donde ocurre el movimiento.

### IV. Aplicación Práctica: Estática y Cadenas

Este desarrollo no solo sirve para partículas en movimiento. En estática, se usa para encontrar la forma de cables bajo tensión o películas de jabón. Cuando minimizas $\sqrt{1 + \dot{f}^2}$, estás buscando la configuración de **mínima energía potencial elástica** o mínima superficie, lo que conecta la mecánica de partículas con la mecánica de medios continuos.

Para una **partícula libre** en mecánica clásica no relativista, el sistema se define por la ausencia de fuerzas externas, lo que implica que la energía potencial es nula ($V = 0$). Este es el escenario más puro para observar la relación entre simetría y leyes de conservación.

Aquí tienes el desarrollo paso a paso:

---

## 1. El Lagrangiano ($L$)

En mecánica lagrangiana, el Lagrangiano se define como la diferencia entre la energía cinética ($T$) y la energía potencial ($V$):


$$L = T - V$$

Para una partícula de masa $m$ que se mueve en el espacio tridimensional con coordenadas cartesianas $(x, y, z)$:

* **Energía cinética ($T$):** $\frac{1}{2}m(\dot{x}^2 + \dot{y}^2 + \dot{z}^2)$
* **Energía potencial ($V$):** $0$ (por definición de partícula libre)

Por lo tanto, el **Lagrangiano de la partícula libre** es:


$$L = \frac{1}{2}m(\dot{x}^2 + \dot{y}^2 + \dot{z}^2)$$

O en notación vectorial, utilizando la velocidad $\mathbf{v}$:


$$L = \frac{1}{2}m\mathbf{v}^2$$

---

## 2. Ecuaciones de Euler-Lagrange

La ecuación de Euler-Lagrange para cada coordenada generalizada $q_i$ es:


$$\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{q}_i} \right) - \frac{\partial L}{\partial q_i} = 0$$

Realicemos el cálculo para la coordenada $x$ (el proceso es idéntico para $y$ y $z$):

1. **Derivada parcial respecto a la posición ($\partial L / \partial x$):**
Como el Lagrangiano solo depende de las velocidades y no de la posición:

$$\frac{\partial L}{\partial x} = 0$$



*(Esto indica que $x$ es una **coordenada cíclica**).*
2. **Derivada parcial respecto a la velocidad ($\partial L / \partial \dot{x}$):**

$$\frac{\partial L}{\partial \dot{x}} = \frac{\partial}{\partial \dot{x}} \left( \frac{1}{2}m\dot{x}^2 \right) = m\dot{x}$$



Este resultado es, por definición, el **momento lineal** $p_x$.
3. **Derivada temporal de la cantidad anterior:**

$$\frac{d}{dt} \left( m\dot{x} \right) = m\ddot{x}$$



**Resultado de la ecuación:**


$$m\ddot{x} = 0$$

Esto se traduce en que la aceleración en los tres ejes es nula ($\ddot{x} = 0, \ddot{y} = 0, \ddot{z} = 0$). Analíticamente, esto recupera la **Primera Ley de Newton**: una partícula libre se mueve con velocidad constante en línea recta.

---

## 3. Cantidades que se conservan

A partir del desarrollo anterior y aplicando el **Teorema de Noether** (que vincula simetrías con leyes de conservación), identificamos las siguientes constantes de movimiento:

### A. Momento Lineal ($\mathbf{p}$)

Vimos que $\frac{\partial L}{\partial q_i} = 0$ para las tres coordenadas. Según la ecuación de Euler-Lagrange, esto implica:


$$\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{q}_i} \right) = 0 \implies p_i = \text{constante}$$

* **Simetría asociada:** Homogeneidad del espacio (invariancia bajo traslaciones espaciales). No importa dónde pongas la partícula, el Lagrangiano es el mismo.

## Enunciado del Problema

Considere un sistema compuesto por dos partículas de masas $m_1$ y $m_2$ que pueden deslizarse libremente sobre una superficie horizontal sin fricción. Ambas masas están conectadas entre sí por un resorte ideal de constante elástica $k$ y longitud natural (en reposo) $l$.

El sistema está restringido a moverse únicamente en una dimensión (eje $x$). No existen fuerzas externas actuando sobre las masas.

**Se solicita:**

1. Definir las coordenadas generalizadas adecuadas.
2. Construir el Lagrangiano del sistema.
3. Obtener las ecuaciones de movimiento de Euler-Lagrange para ambas masas.

---

## 1. Coordenadas Generalizadas

Sea $x_1$ la posición de la masa $m_1$ y $x_2$ la posición de la masa $m_2$ respecto a un origen fijo en el laboratorio. La extensión (o compresión) del resorte en cualquier instante viene dada por la diferencia de sus posiciones menos su longitud natural:


$$\Delta x = (x_2 - x_1) - l$$

Nuestras coordenadas generalizadas serán $q_1 = x_1$ y $q_2 = x_2$.

---

## 2. Construcción del Lagrangiano ($L$)

El Lagrangiano es $L = T - V$.

* **Energía Cinética ($T$):** Es la suma de las energías cinéticas individuales.

$$T = \frac{1}{2}m_1\dot{x}_1^2 + \frac{1}{2}m_2\dot{x}_2^2$$


* **Energía Potencial ($V$):** Es la energía elástica almacenada en el resorte debido a su deformación.

$$V = \frac{1}{2}k(x_2 - x_1 - l)^2$$



Por lo tanto:


$$L = \frac{1}{2}m_1\dot{x}_1^2 + \frac{1}{2}m_2\dot{x}_2^2 - \frac{1}{2}k(x_2 - x_1 - l)^2$$

---

## 3. Obtención de las Ecuaciones de Lagrange

Aplicamos la ecuación de Euler-Lagrange para cada coordenada: $\frac{d}{dt} \left( \frac{\partial L}{\partial \dot{x}_i} \right) - \frac{\partial L}{\partial x_i} = 0$.

### Para la masa $m_1$:

1. **Derivada respecto a la velocidad:** $\frac{\partial L}{\partial \dot{x}_1} = m_1\dot{x}_1$.
2. **Derivada temporal:** $\frac{d}{dt}(m_1\dot{x}_1) = m_1\ddot{x}_1$.
3. **Derivada respecto a la posición:**
Usando la regla de la cadena en el término de $V$:

$$\frac{\partial L}{\partial x_1} = -\frac{\partial V}{\partial x_1} = -k(x_2 - x_1 - l) \cdot (-1) = k(x_2 - x_1 - l)$$



**Ecuación 1:**


$$m_1\ddot{x}_1 - k(x_2 - x_1 - l) = 0 \implies m_1\ddot{x}_1 = k(x_2 - x_1 - l)$$

---

### Para la masa $m_2$:

1. **Derivada respecto a la velocidad:** $\frac{\partial L}{\partial \dot{x}_2} = m_2\dot{x}_2$.
2. **Derivada temporal:** $\frac{d}{dt}(m_2\dot{x}_2) = m_2\ddot{x}_2$.
3. **Derivada respecto a la posición:**

$$\frac{\partial L}{\partial x_2} = -\frac{\partial V}{\partial x_2} = -k(x_2 - x_1 - l) \cdot (1) = -k(x_2 - x_1 - l)$$



**Ecuación 2:**


$$m_2\ddot{x}_2 + k(x_2 - x_1 - l) = 0 \implies m_2\ddot{x}_2 = -k(x_2 - x_1 - l)$$

---

## Resumen y Análisis

Las ecuaciones de Lagrange resultantes son:

$$\begin{cases} 
m_1\ddot{x}_1 = k(x_2 - x_1 - l) \\
m_2\ddot{x}_2 = -k(x_2 - x_1 - l) 
\end{cases}$$

**Interpretación Física:**

* Nota que las fuerzas son iguales y opuestas (tercera ley de Newton). El resorte "tira" de la masa 1 hacia la derecha (término positivo) y de la masa 2 hacia la izquierda (término negativo) cuando está estirado.
* Si sumas ambas ecuaciones, obtienes $m_1\ddot{x}_1 + m_2\ddot{x}_2 = 0$. Esto implica que la aceleración del **Centro de Masas** es cero, lo cual es consistente con el hecho de que no hay fuerzas externas actuando sobre el sistema (conservación del momento lineal total).
* Para resolver el movimiento relativo (la oscilación), se suele definir una coordenada relativa $u = x_2 - x_1 - l$, lo que reduciría el sistema a una sola ecuación de oscilador armónico con la **masa reducida** $\mu = \frac{m_1 m_2}{m_1 + m_2}$.

### 1. Movimiento del Centro de Masas ($R$)

Sumemos las dos ecuaciones:


$$m_1\ddot{x}_1 + m_2\ddot{x}_2 = k(x_2 - x_1 - l) - k(x_2 - x_1 - l) = 0$$

Definimos la posición del centro de masas como $R = \frac{m_1x_1 + m_2x_2}{m_1 + m_2}$. La ecuación anterior nos dice que:


$$(m_1 + m_2)\ddot{R} = 0 \implies \ddot{R} = 0$$


**Solución:** El centro de masas se mueve como una partícula libre (velocidad constante):


$$R(t) = V_0 t + R_0$$

---

### 2. Movimiento Relativo ($r$)

Ahora queremos saber cómo cambia la distancia entre ellas. Definamos la coordenada relativa como la deformación del resorte:


$$u = x_2 - x_1 - l$$


Derivamos dos veces respecto al tiempo: $\ddot{u} = \ddot{x}_2 - \ddot{x}_1$.

Sustituimos las aceleraciones de las ecuaciones originales:


$$\ddot{u} = \left( -\frac{k}{m_2}u \right) - \left( \frac{k}{m_1}u \right) = -k \left( \frac{1}{m_1} + \frac{1}{m_2} \right) u$$

Aquí aparece un concepto fundamental, la **masa reducida ($\mu$)**:


$$\frac{1}{\mu} = \frac{1}{m_1} + \frac{1}{m_2} \implies \mu = \frac{m_1 m_2}{m_1 + m_2}$$

La ecuación se convierte en un **oscilador armónico simple**:


$$\ddot{u} + \frac{k}{\mu} u = 0$$


**Solución:** Definimos la frecuencia angular $\omega = \sqrt{k/\mu}$:


$$u(t) = A \cos(\omega t + \phi)$$

---

### 3. Solución Final para $x_1$ y $x_2$

Una vez que tienes $R(t)$ y $u(t)$, simplemente haces un cambio de variables inverso. Sabemos por definición que:

1. $x_2 - x_1 = u(t) + l$
2. $m_1x_1 + m_2x_2 = (m_1 + m_2)R(t)$

Resolviendo este sistema de 2x2 para $x_1$ y $x_2$:

* $x_1(t) = R(t) - \frac{m_2}{m_1 + m_2}(u(t) + l)$
* $x_2(t) = R(t) + \frac{m_1}{m_1 + m_2}(u(t) + l)$

### Conclusión

¿Es fácil? **Sí**, siempre y cuando no intentes resolver las ecuaciones acopladas directamente. Al pasar a coordenadas del centro de masas y coordenadas relativas, conviertes un problema complejo de dos cuerpos en dos problemas simples de un solo cuerpo:

1. Una partícula que se mueve en línea recta (Inercia).
2. Una partícula que oscila respecto a un punto fijo (Armónico).
