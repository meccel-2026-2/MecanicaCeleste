```python
import numpy as np
import pymcel as pc
import matplotlib.pyplot as plt
```

```python
sistema = [
    # Particula 1
    dict(
    m=100.0,
    r=[-0.142857,0,0],
    v=[0,-2.958,0]),

    dict(
    m = 5.0,
    r = [2.85714,0,0],
    v = [0,2.958,0]
    ),

    dict(
    m = 0.01,
    r = [3.62345,0,0],
    v = [0,4.9700107,0]
    )
    ]
```

```python
ts = np.linspace(0,6.5,100)
rs, vs, rps, vps, cuadraturas = pc.ncuerpos_solucion(sistema, ts)
```

```python
plt.plot(rps[0,:,0],rps[0,:,1], "k-", label = "Particula 1")
plt.plot(rps[1,:,0],rps[1,:,1], "b-", label = "Particula 2")
plt.plot(rps[2,:,0],rps[2,:,1], "r-", label = "Particula 3")
plt.axis("equal")
plt.legend();
```

Sistema jerárquico:

[(P1,P2),P3]

Este sistema es anomalo

Vamos a hacer una animación

```python
from celluloid import Camera
```

```python
fig = plt.figure()
camera = Camera(fig)

for i, t in enumerate(ts[::10]):
    plt.plot(rps[0,i,0],rps[0,i,1], "ko", label = "Particula 1")
    plt.plot(rps[1,i,0],rps[1,i,1], "bo", label = "Particula 2")
    plt.plot(rps[2,i,0],rps[2,i,1], "ro", label = "Particula 3")
    camera.snap()

plt.grid(ls="--")
plt.axis("equal");
```

```python
from IPython.display import HTML

anim = camera.animate()
HTML(anim.to_jshtml())
```
